## Задание 1
Посчитайте сумму всех целых чисел от 1 до 100.

## Задание 3
Посчитайте, сколько раз встречается буква "а" в предложении.

````python
sentence = "съешь ещё этих мягких французских булок, да выпей же чаю"
count_a = 0
for ch in sentence:
    if ch == 'а':
        count_a += 1

print(count_a)
````

# Задание 4
Посчитайте, сколько слов в списке имеют корень **"хорош"**.

````python
words = [
    "великолепный", "хорошенький", "скверный", "приятный", "слабый",
    "хороший", "уродливый", "замечательный", "тусклый", "достойный",
    "ужасный", "яркий", "хорошенечкий", "посредственный", "красивый",
    "прекрасный", "сильный", "неплохой", "отличный", "хорошеватый",
    "приличный", "восхитительный", "плохой", "потрясающий"
]

count = 0
for word in words:
    if "хорош" in word:   # проверка, содержится ли подстрока
        count += 1

print(count)
````

### Срезы
* Переменная[номер]
* Переменная[от:до], до - не включительно
* Переменная[от:до:шаг]

```
word = "клаустрофобия"
#       0123456789...

print(word[2])    # а
print(word[2:3])  # а
print(word[:3])   # кла
print(word[1:3])  # ла
print(word[1:6:2])  # лут
print(word[-3:]) # бия
print(word[55])  # 🛑
```

# Задания 5
В этом же списке посчитайте, сколько слов начинается с **"пр"**, как **пр**ивет

````python
words = [
    "великолепный", "хорошенький", "скверный", "приятный", "слабый",
    "хороший", "уродливый", "замечательный", "тусклый", "достойный",
    "ужасный", "яркий", "хорошенечкий", "посредственный", "красивый",
    "прекрасный", "сильный", "неплохой", "отличный", "хорошеватый",
    "приличный", "восхитительный", "плохой", "потрясающий"
]

count = 0
for word in words:
    if word.startswith("пр"):   # метод строк startswith
        count += 1

print(count)
````

# Задание 6 (бонусное)
И напоследок посчитайте, сколько прилагательных в этом списке заканчивается на "-ый", сколько - на "-ий", и сколько - на другие окончания

````python
words = [
    "великолепный", "хорошенький", "скверный", "приятный", "слабый",
    "хороший", "уродливый", "замечательный", "тусклый", "достойный",
    "ужасный", "яркий", "хорошенечкий", "посредственный", "красивый",
    "прекрасный", "сильный", "неплохой", "отличный", "хорошеватый",
    "приличный", "восхитительный", "плохой", "потрясающий"
]

count_iy = 0
count_iy2 = 0
count_other = 0

for word in words:
    last_two = word[-2:]
    if last_two == "ый":
        count_iy += 1
    elif last_two == "ий":
        count_iy2 += 1
    else:
        count_other += 1

print("-ый:", count_iy)
print("-ий:", count_iy2)
print("другие окончания:", count_other)
````
